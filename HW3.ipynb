```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston.keys() # просмотр содержимого датасета BOSTON HOUSE Prices
```

```python
data = boston.data # Данные о недвижимости ()(значения различных наблюдений о них)
features = boston.feature_names # Признаки
target = boston.target # Целева япеременная (цена на недвижимость)
```

```python
X = pd.DataFrame(data, columns = features) # Для оценик целевой перемнной выбираем значения признаков, созданных на основание данных data
Y = pd.DataFrame(target,columns = (['price']))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Разбиение выборки на тестовую и тренировчную
x_test,x_train,y_test,y_train = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```
