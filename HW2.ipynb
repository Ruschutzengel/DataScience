### Задание 2
С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

- Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.
- Выведите название диаграммы - 'График f(t)'. 
- Также добавьте названия для горизонтальной оси -'Значения t' и для вертикальной - 'Значения f'.
- Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0,10,num = 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
title_font = {
    'fontsize': 16,
    'fontweight': 10,
    'color': '#191970',
    'family':'Times New Roman'}

label_font = {
    'fontsize': 12,
    'fontweight': 8,
    'color': '#191970',
    'family':'Times New Roman'}

plt.plot(t,f, label ='cos(x)', color ='green')

plt.title('График f(t)',fontdict = title_font)
plt.xlabel('x', loc ='center', fontdict =label_font)
plt.ylabel('cos(x)', loc ='center', fontdict =label_font)

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')

DATASET_PATH = './creditcard.csv'
```

```python
cr = pd.read_csv(DATASET_PATH, sep =',')
cr.describe()
```

Для понимания параметров из файла:
- Time: Number of seconds elapsed between this transaction and the first transaction in the dataset
- Amount: transaction amount
- Class: 1 for fraudulent transactions, 0 otherwise
- [V1 - V28]: may be result of a PCA Dimensionality reduction to protect user identities and sensitive features(v1-v28)

```python
print('1. shape: {}'.format(cr.shape))
print('2. index: {}'.format(cr.index))
print('3. size: {}'.format(cr.size))
print('4. Columns:\n {}'.format(cr.columns))
```

```python
cr1 = ( cr[['V1','Class']] ) # Срез из набора данных
print(cr1)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

fig.subplots_adjust(hspace = 0.3, wspace = 0.3) # Вертикальные зазоры графика в дюймах
rc['figure.figsize'] = (8,6) #Установка размера фигуры 8' x 6'

y1 = cr1.query('Class == 0')
y2 = cr1.query('Class == 1')

ax1,ax2 = ax.flatten() # Преобразование в одномерный массив,чтобы не брать кортежи

ax1.legend('Class 0')
ax2.legend('Class 1')

ax1.hist(y1, bins = 20, density = True, alpha = 0.5) # Гистограмма для обычных транзакций Class = 0
ax2.hist(y2, bins = 20,density =True, alpha = 0.5) # Гистограмма для мошеннеческих транзакций Class = 1
```
